#**Reduction Operation**

```python
#mean
torch.mean(e)
#mean along col
torch.mean(e, dim=0)
```

```python
#median
torch.median(e)
```

```python
#max and min
torch.max(e)
torch.min(e)
```

```python
#product
torch.prod(e)
```

```python
#standard deviation
torch.std(e)
```

```python
#variance
torch.var(e)
```

```python
torch.argmax(e) #argmax
```

```python
#argmin
torch.argmin(e)
```

#**Matrix Operations**

```python
f = torch.randint(size=(2,3), low=0, high=10)
g = torch.randint(size=(3,2), low=0, high=10)

print(f)
print(g)
```

```python
#matrix multiplication
torch.matmul(f, g)
```

```python
vector1 = torch.tensor([1,2])
vector2 = torch.tensor([3,4])

#dot product
torch.dot(vector1, vector2)
```

#**Comparison Operations**

```python
i = torch.randint(size=(2,3), low=0, high=10)
j = torch.randint(size=(2,3), low=0, high=10)

print(i)
print(j)
```

```python
#greater than
i > j
#less than
i < j
#equal to
i == j
#not equal to
i != j
```
